```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas.util.testing as tm
import seaborn as sns
%matplotlib inline
```

```
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/teleCust1000t.csv"
data = pd.read_csv(url)
```

```
data['age'] = data['age'].astype('float64')
data['custcat'] = data['custcat'].astype('float64')
data['region'] = data['region'].astype('float64')
```

```
data.dtypes
```

```
from sklearn.model_selection import train_test_split
```

```
X = data[['income','age','region']]
y= data['custcat']

```

```
print(X.shape)
print(y.shape)
```

```
from sklearn.model_selection import train_test_split

X_train , X_test, y_train,y_test = train_test_split(X,y,test_size=0.4)
```

```
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
model = KNeighborsClassifier(n_neighbors=4)
```

```
model.fit(X_train, y_train)
```

```
prediction = model.predict(X_test)
```

```
print(prediction)
```

```
print(data['custcat'])
```

```
from sklearn import metrics
```

```
print('Test set accuracy : ', metrics.accuracy_score(y_test,prediction))
```

```
X = data[['tenure','age','address','income','employ']]
y = data['custcat']
```

```
print(X.shape)
print(y.shape)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4,random_state=0)
```

```
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
knn = KNeighborsClassifier(n_neighbors=4)
```

```
knn.fit(X_train,y_train)
```

```
knn_pred = knn.predict(X_test)
```

```
print(knn_pred)
```

```
from sklearn import metrics
```

```
print('Test accuracy : ', metrics.accuracy_score(y_test,knn_pred))
```

```
print(metrics.classification_report(y_test,knn_pred))
```

```
knn1 = KNeighborsClassifier(n_neighbors=5)
knn1.fit(X_train,y_train)
knn1_pred = knn1.predict(X_test)
```
